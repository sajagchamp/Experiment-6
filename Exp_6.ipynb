```python
#6.6 check whether the given year is leap year or not :

year=int(input("enter a year:"))
if year%4==0 and year%100!=0:
  print("year is leap year")
elif year%400==0 and year%100==0:
  print("year is leap year")
else:
  print("year is not leap year")
```

```python
#6.9 Salary Calculation with Allowances.
# Write a python program to calculate the gross salary of an employee.
# If basic<=10000: HRA = 20%, DA = 80%.
#If basic <= 20000: HRA = 25%, DA = 90%.
#Otherwise: HA = 30%, DA=95%.
basic=float(input("Enter basic salary: "))
if basic<=10000:
   hra=basic * 0.20
   da = basic * 0.80
elif basic <= 20000:
   hra= basic * 0.25
   da=basic * 0.90
else:
   hra= basic * 0.30
   da= basic * 0.95
gross= basic+hra+da

print("Basic Salary:", basic)
print("HRA:", hra)
print("DA:", da)
print("Gross Salary:", gross)
```

```python
#6.10 Write a python program to calculate income tax based on annual income:
# Up to Rs.2,50,000: No tax
# Rs.2,50,000-Rs.5,00,000: 5%
# Rs.5,00,001- Rs.10,00,000: 20%
#Above Rs.10,00,000: 30%
income = float(input("Enter your annual income: "))
if income <= 250000:
  tax = 0
  print("No income tax to be paid.")
elif income <= 500000:
    tax = (income-250000) * 0.05
    print("Income tax to be paid:", tax)
elif income <= 1000000:
    tax = (250000*0.05) + (income-500000) * 0.20
    print("Income tax to be paid:", tax)
else :
    tax = (250000 * 0.05) + (500000 * 0.20) + (income-1000000) * 0.30
    print("Income tax to be paid:", tax)
```
